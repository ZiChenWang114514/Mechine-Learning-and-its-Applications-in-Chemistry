#### 2.3 ç”Ÿæˆ2Då’Œ3Dæ„è±¡ <a id='2-3'></a>  
**2Dæ„è±¡**

åœ¨å‰é¢å­¦ä¹ â€œç»˜åˆ¶å¤šä¸ªåˆ†å­â€æ—¶ï¼Œæˆ‘ä»¬å‘ç°åˆ†å­ç»“æ„å±•ç¤ºæ•ˆæœä¸æ˜¯å¾ˆå¥½ï¼Œç°åœ¨å¯ç”¨AllChem.Compute2DCoords()è§£å†³è¿™ä¸ªé—®é¢˜ã€‚

```python
from rdkit.Chem import AllChem
suppl = Chem.SDMolSupplier('RDKit-data/ligands.sdf')
mols = [x for x in suppl if x is not None ]

# é‡æ–°è®¡ç®—2Dæ„è±¡
for mol in mols:
    AllChem.Compute2DCoords(mol)
Chem.Draw.MolsToImage(mols,molsPerRow=4,subImgSize=(200,200),legends=[x.GetProp("_Name") for x in mols])    
```

**3Dæ„è±¡**  
RDKit å¯ä»¥é€šè¿‡ä¸¤ç§æ–¹æ³•ä¸ºåŒ–åˆç‰©ç”Ÿæˆ3Dæ„è±¡ã€‚
1. å‡ ä½•è·ç¦»æ³•ç”Ÿæˆåˆå§‹æ„è±¡ï¼Œç„¶åç”¨åŠ›åœºUFFæˆ–è€…MMFFä¼˜åŒ–æ„è±¡ï¼›
2. ETKDGæ–¹æ³•ï¼Œå®ƒä½¿ç”¨æ¥è‡ªå‰‘æ¡¥ç»“æ„æ•°æ®åº“çš„æ‰­è½¬è§’ä¿¡æ¯å¯¹ç”Ÿæˆçš„æ„è±¡è¿›è¡Œæ ¡æ­£ã€‚  

<font color=DarkGoldenRod>å€¼å¾—æ³¨æ„çš„æ˜¯ï¼Œæ„è±¡ç”Ÿæˆæ˜¯ä¸€é¡¹å›°éš¾ä¸”ç»†å¾®çš„å·¥ä½œï¼ŒRDKitä¸ºä½¿ç”¨è€…æä¾›äº†ä¸€ä¸ªè¾ƒä¸ºå¿«é€Ÿçš„3Dæ„è±¡ç”Ÿæˆæ–¹æ³•ï¼Œä½†è¿™ç§æ–¹æ³•å¹¶ä¸èƒ½å–ä»£çœŸæ­£çš„æ„è±¡ç”Ÿæˆç¨‹åºã€‚</font> 

è¿™é‡Œæ˜¯ä¸€ä¸ªä½¿ç”¨MMFF94é€šè¿‡èƒ½é‡æœ€å°åŒ–ç”ŸæˆåŒ–åˆç‰©3Dæ„è±¡çš„ä¾‹å­ï¼š

```python
mols2 = []
for mol in mols:
    AllChem.Compute2DCoords(mol)
    mol2=Chem.AddHs(mol)
    AllChem.EmbedMolecule(mol2)
    AllChem.MMFFOptimizeMolecule(mol2)
    mols2.append(mol2)
    
Draw.MolsToImage(mols2,molsPerRow=4,subImgSize=(200,200),legends=[x.GetProp("_Name") for x in mols])    
```

RDKitä¹Ÿå¯ä»¥ä¸ºåŒ–åˆç‰©ç”Ÿæˆå¤šç§æ„è±¡ï¼Œæˆ‘ä»¬éœ€è¦é€šè¿‡numConfsè®¾ç½®æƒ³è¦çš„æ„è±¡æ•°é‡ã€‚  

```python
m = Chem.MolFromSmiles('C1CCC1OC')
m2=Chem.AddHs(m)
# run ETKDG 10 times
cids = AllChem.EmbedMultipleConfs(m2, numConfs=10)
print(len(cids))

rmslist = []
#ä½¿ç”¨AllChem.AlignMolConformers(),è®¡ç®—ä¸åŒæ„è±¡é—´çš„RMSDå·®å¼‚ã€‚
AllChem.AlignMolConformers(m2, RMSlist=rmslist)
print(len(rmslist))
```

æŠŠnumThreadsè®¾ä¸º0æ—¶ï¼Œç¨‹åºä¼šä½¿ç”¨è®¡ç®—æœºæ‰€å…è®¸çš„æ‰€æœ‰çº¿ç¨‹ã€‚

```python
cids = AllChem.EmbedMultipleConfs(m2, numThreads=0)
res = AllChem.MMFFOptimizeMoleculeConfs(m2, numThreads=0)
```

#### 2.6 åˆ†å­æŒ‡çº¹ä¸åˆ†å­ç›¸ä¼¼æ€§   <a id='2-6'></a>  
åœ¨åŒ–å­¦ä¿¡æ¯å­¦çš„é¢†åŸŸä¸­ï¼Œåˆ†å­æŒ‡çº¹æ˜¯ä¸€é¡¹é‡è¦çš„æŠ€æœ¯ï¼Œç”¨äºå°†åˆ†å­ç»“æ„è½¬åŒ–ä¸ºæ•°å€¼è¡¨ç¤ºï¼Œä»¥ä¾¿è¿›è¡Œç›¸ä¼¼æ€§æ¯”è¾ƒã€æœºå™¨å­¦ä¹ å’Œè¯ç‰©å‘ç°ç­‰ä»»åŠ¡ã€‚  
RDKitæ”¯æŒå¤šç§åˆ†å­æŒ‡çº¹ï¼ŒåŒ…æ‹¬Morganã€MACCSã€åŸå­å¯¹ã€æ‹“æ‰‘æ‰­è½¬è§’ã€æ‹“æ‰‘æŒ‡çº¹ç­‰ã€‚ 

è·å–æŒ‡çº¹æœ€ç›´æ¥çš„æ–¹æ³•æ˜¯ä¸ºæ„Ÿå…´è¶£çš„æŒ‡çº¹ç±»å‹åˆ›å»ºä¸€ä¸ªFingeprintGeneratorå¯¹è±¡ï¼Œç„¶åä½¿ç”¨å®ƒæ¥è®¡ç®—æŒ‡çº¹ã€‚  

**MorganæŒ‡çº¹**   
MorganæŒ‡çº¹æ˜¯å¸¸ç”¨çš„åˆ†å­æŒ‡çº¹ä¹‹ä¸€ï¼Œå®ƒåŸºäºåˆ†å­ä¸­çš„ç¯å½¢å­ç»“æ„æ¥æè¿°åˆ†å­çš„æ‹“æ‰‘ç»“æ„ç‰¹å¾ã€‚

MorganæŒ‡çº¹çš„ç”Ÿæˆè¿‡ç¨‹ï¼š
1. åŠå¾„åˆ†é…ã€‚ä¸ºåˆ†å­ä¸­çš„æ¯ä¸ªåŸå­åˆ†é…ä¸€ä¸ªåŠå¾„ï¼Œä»¥ä¾¿ç¡®å®šåˆ†å­ä¸­å„ä¸ªåŸå­çš„é‚»å±…å…³ç³»ã€‚
2. ç¯å½¢å­ç»“æ„ç”Ÿæˆã€‚åœ¨æ¯ä¸ªåŠå¾„å±‚æ¬¡ä¸Šï¼Œé€šè¿‡éå†åˆ†å­çš„åŸå­å’Œé”®çš„å…³ç³»ï¼Œæˆ‘ä»¬å¯ä»¥æ‰¾åˆ°åˆ†å­ä¸­çš„æ‰€æœ‰ç¯å½¢å­ç»“æ„ã€‚
3. ä½¿ç”¨å“ˆå¸Œå‡½æ•°å¯¹è¿™äº›ç¯å½¢å­ç»“æ„è¿›è¡Œç¼–ç ï¼Œå¹¶å°†å…¶æ˜ å°„ä¸ºäºŒè¿›åˆ¶ä½ä¸²ã€‚

```python
from rdkit.Chem import AllChem
from rdkit import DataStructs
fpgen = AllChem.GetMorganGenerator(radius=2)
m1 = Chem.MolFromSmiles('Cc1ccccc1')
m2 = Chem.MolFromSmiles('Cc1ncccc1')
fp1 = fpgen.GetSparseCountFingerprint(m1)
fp2 = fpgen.GetSparseCountFingerprint(m2)
DataStructs.DiceSimilarity(fp1,fp2)
```

**MACCSæŒ‡çº¹**  
MACCSæŒ‡çº¹æ˜¯å¦ä¸€ç§å¸¸ç”¨çš„åˆ†å­æŒ‡çº¹ç±»å‹ï¼Œå®ƒåŸºäºåˆ†å­ä¸­çš„åŒ–å­¦åŠŸèƒ½å›¢å’Œå­ç»“æ„æ¥æè¿°åˆ†å­çš„ç»“æ„ç‰¹å¾ã€‚ 

MACCSæŒ‡çº¹çš„ç”Ÿæˆè¿‡ç¨‹ï¼š
1. å®šä¹‰äº†ä¸€ç»„SMARTSæ¨¡å¼ï¼Œè¿™äº›æ¨¡å¼æè¿°äº†å¸¸è§çš„åŒ–å­¦åŠŸèƒ½å›¢å’Œå­ç»“æ„ã€‚
2. å¯¹äºç»™å®šçš„åˆ†å­ï¼Œæˆ‘ä»¬é€ä¸ªåŒ¹é…è¿™äº›SMARTSæ¨¡å¼ï¼Œå¦‚æœåˆ†å­ä¸­å­˜åœ¨ä¸æ¨¡å¼åŒ¹é…çš„å­ç»“æ„ï¼Œåˆ™å°†ç›¸åº”çš„ä½è®¾ç½®ä¸º1ï¼Œå¦åˆ™ä¸º0ã€‚
3. å°†æ‰€æœ‰å­ç»“æ„æ¨¡å¼çš„åŒ¹é…ç»“æœåˆå¹¶ä¸ºä¸€ä¸ªMACCSæŒ‡çº¹ã€‚

```python
from rdkit.Chem import MACCSkeys

m1 = Chem.MolFromSmiles('Cc1ccccc1')
m2 = Chem.MolFromSmiles('Cc1ncccc1')
fp1 = MACCSkeys.GenMACCSKeys(m1)
fp2 = MACCSkeys.GenMACCSKeys(m2)
DataStructs.DiceSimilarity(fp1,fp2)
```

##### **Atom PairæŒ‡çº¹**  
Atom PairæŒ‡çº¹æ˜¯ä¸€ç§åŸºäºåˆ†å­ä¸­åŸå­å¯¹ï¼ˆatom pairï¼‰çš„ç‰¹å¾è¡¨ç¤ºæ–¹æ³•ã€‚å®ƒé€šè¿‡è®¡ç®—åˆ†å­ä¸­æ‰€æœ‰åŸå­å¯¹ä¹‹é—´çš„è·ç¦»ï¼Œå¹¶å°†è¿™äº›è·ç¦»è½¬åŒ–ä¸ºäºŒè¿›åˆ¶ä½ä¸²æ¥è¡¨ç¤ºåˆ†å­çš„ç»“æ„ç‰¹å¾ã€‚Atom PairæŒ‡çº¹å¯ä»¥æ•æ‰åŸå­ä¹‹é—´çš„ç›¸äº’ä½œç”¨å’Œç©ºé—´å…³ç³»ï¼Œå› æ­¤åœ¨æè¿°åˆ†å­çš„ç«‹ä½“ç»“æ„å’Œæ€§è´¨æ—¶éå¸¸æœ‰ç”¨ã€‚

Atom PairæŒ‡çº¹çš„ç”Ÿæˆè¿‡ç¨‹å¦‚ä¸‹ï¼š
1. è®¡ç®—åˆ†å­ä¸­æ‰€æœ‰åŸå­å¯¹ä¹‹é—´çš„è·ç¦»ï¼Œå¹¶å°†å…¶æ˜ å°„åˆ°ä¸€ä¸ªå›ºå®šçš„èŒƒå›´ã€‚
2. å°†è¿™äº›è·ç¦»åˆ†é…ç»™å¯¹åº”çš„ä½ï¼Œå¦‚æœæŸä¸ªåŸå­å¯¹çš„è·ç¦»è½åœ¨æŸä¸ªèŒƒå›´å†…ï¼Œåˆ™å°†å¯¹åº”ä½è®¾ç½®ä¸º1ï¼Œå¦åˆ™ä¸º0ã€‚
3. å°†æ‰€æœ‰ä½çš„çŠ¶æ€åˆå¹¶ä¸ºä¸€ä¸ªAtom PairæŒ‡çº¹ã€‚

```python
from rdkit.Chem import AllChem
fpgen = AllChem.GetRDKitFPGenerator(maxPath=2,fpSize=1024)
m1 = Chem.MolFromSmiles('Cc1ccccc1')
m2 = Chem.MolFromSmiles('Cc1ncccc1')
fp1 = fpgen.GetSparseCountFingerprint(m1)
fp2 = fpgen.GetSparseCountFingerprint(m2)
DataStructs.DiceSimilarity(fp1,fp2)
```

**Topological TorsionæŒ‡çº¹**  
Topological TorsionæŒ‡çº¹åŸºäºåˆ†å­ä¸­çš„æ‹“æ‰‘æ‰­æ›²ç‰¹å¾æ¥æè¿°åˆ†å­çš„ç»“æ„ã€‚å®ƒä¸»è¦ç”¨äºæ•æ‰åˆ†å­ä¸­çš„ç¯å½¢å­ç»“æ„å’Œæ‰­æ›²æ€§è´¨ï¼Œå¯¹äºæè¿°åˆ†å­çš„ç«‹ä½“åŒ–å­¦ç‰¹å¾å…·æœ‰é‡è¦æ„ä¹‰ã€‚

Topological TorsionæŒ‡çº¹çš„ç”Ÿæˆè¿‡ç¨‹å¦‚ä¸‹ï¼š
1. è¯†åˆ«åˆ†å­ä¸­çš„æ‰€æœ‰ç¯å½¢å­ç»“æ„ï¼Œå¹¶ç¡®å®šè¿™äº›ç¯å½¢å­ç»“æ„çš„æ‰­æ›²ç¨‹åº¦ã€‚ç¯çš„æ‰­æ›²ç¨‹åº¦æ˜¯é€šè¿‡è®¡ç®—ç¯ä¸­çš„åŒ–å­¦é”®çš„æ—‹è½¬è§’åº¦æ¥è¡¡é‡çš„ã€‚
2. å°†è¿™äº›æ‰­æ›²ç¨‹åº¦æ˜ å°„åˆ°ä¸€ä¸ªå›ºå®šçš„èŒƒå›´ï¼Œå¹¶å°†å…¶åˆ†é…ç»™å¯¹åº”çš„ä½ã€‚å¦‚æœæŸä¸ªç¯å½¢å­ç»“æ„çš„æ‰­æ›²ç¨‹åº¦è½åœ¨æŸä¸ªèŒƒå›´å†…ï¼Œåˆ™å°†å¯¹åº”ä½è®¾ç½®ä¸º1ï¼Œå¦åˆ™ä¸º0ã€‚
3. å°†æ‰€æœ‰ä½çš„çŠ¶æ€åˆå¹¶ä¸ºä¸€ä¸ªTopological TorsionæŒ‡çº¹ã€‚

```python
from rdkit.Chem import AllChem

pgen = AllChem.GetTopologicalTorsionGenerator()
m1 = Chem.MolFromSmiles('Cc1ccccc1')
m2 = Chem.MolFromSmiles('Cc1ncccc1')
fp1 = fpgen.GetSparseCountFingerprint(m1)
fp2 = fpgen.GetSparseCountFingerprint(m2)
DataStructs.DiceSimilarity(fp1,fp2)
```
